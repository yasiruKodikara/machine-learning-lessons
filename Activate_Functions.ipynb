# sigmoid function


```python
import numpy as np
#sigmoid activation function

def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))

inputs = [13,3,-12]
outputs = []

for x in inputs:
  print(sigmoid(x))
  outputs.append(sigmoid(x))

import matplotlib.pyplot as plt
plt.plot(inputs, outputs)
plt.show()
```

# Tanh Function


```python
import numpy as np
import matplotlib.pyplot as plt

#tanh activation function
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

inputs = [13,3,-12]
outputs = []

for x in inputs:
  print(tanh(x))
  outputs.append(tanh(x))

plt.plot(inputs, outputs)
plt.show()
```

# ReLU Function

```python
def relu(x):
  return max(0, x)

inputs = [13,3,-12]
outputs = []

for x in inputs:
  print(relu(x))
  outputs.append(relu(x))

plt.plot(inputs, outputs)
plt.show()
```

# Leaky ReLU Function


```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)

inputs = [13,3,-12]
outputs = []

for x in inputs:
  print(softmax(x))
  outputs.append(softmax(x))

plt.plot(inputs, outputs)
plt.show()
```

| Function   | Input Cases     | Output Examples                                   | Usage                                   |
|------------|-----------------|---------------------------------------------------|-----------------------------------------|
| Sigmoid    | Positive, Zero, Negative | ~1 (positive), ~0.95 (positive), ~0 (negative) | Good for binary classification outputs. Squashes values between 0 and 1. |
| Tanh       | Positive, Zero, Negative | ~1 (positive), ~1 (positive), ~-1 (negative) | Squashes values between -1 and 1. Often used in hidden layers. |
| ReLU       | Positive, Zero, Negative | Positive, Positive, Zero                  | Simple and effective. Outputs 0 for negative inputs and the input itself for positive inputs. |
| Softmax    | Positive, Zero, Negative | 1.0, 1.0, 1.0                             | Used for multi-class classification outputs. Converts outputs to probability distributions. (Note: The current code applies Softmax element-wise, which is not the typical usage for multi-class classification. Typically, Softmax is applied to a vector of outputs.) |
